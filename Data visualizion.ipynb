```python
# importing relevant libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the dataset into a pandas DataFrame
df = pd.read_csv("C:\\Users\\ASUS\\Desktop\\PROJECTS_DS\\Accenture-virtual-internship\\Task 3_Final Content Data set.csv")
df.head()
```

```python
# Bar Chart for Top 5 Content Categories
top_5_categories = df['Category'].value_counts().head(5)
plt.bar(top_5_categories.index, top_5_categories.values)
plt.xlabel('Content Category')
plt.ylabel('Number of Posts')
plt.title('Top 5 Content Categories')
plt.show()
```

```python
category_counts = df['Category'].value_counts()

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(category_counts.values, labels=category_counts.index, autopct='%1.1f%%')
plt.title('Distribution of Different Categories')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
top_5_categories = df['Category'].value_counts().head(5).index.tolist()
df_top_5_categories = df[df['Category'].isin(top_5_categories)]
```

```python
reaction_c = df_top_5_categories.groupby(['Category', 'Reaction Type']).size().unstack()
reaction_c.plot(kind='bar', stacked=True, figsize=(20, 10))
plt.xlabel('Categories')
plt.ylabel('Reaction Type Count')
plt.title('Reaction Type Count for Top 5 Categories')
plt.xticks(rotation=45)  # Rotate the x-axis labels for better visibility
plt.legend(title='Reaction Type')
plt.tight_layout()
plt.show()

```

```python
# Line Chart for Posts Over Time 
df['Datetime'] = pd.to_datetime(df['Datetime'])
df['month'] = df['Datetime'].dt.month
posts_per_month = df['month'].value_counts().sort_index()
plt.plot(posts_per_month.index, posts_per_month.values)
plt.xlabel('Month')
plt.ylabel('Number of Posts')
plt.title('Number of Posts Over Time')
plt.show()
```
